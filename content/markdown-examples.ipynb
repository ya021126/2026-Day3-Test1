# Markdown examples

To write formatted text in notebooks, use "markdown cells". Go to cell type (defaults to "Code") just above fast-forward sign in notebook menu above, and select "Markdown".

Then "play" will render the markdown into something that looks nice.


## Simple markdown!

**Bold** and *italic* with * and **.

Links with square brackets and round brackets:

`[Link to google](https://google.com/)` renders to:

[Link to google](https://google.com/)

## Headings and lists

The number of hashes in front of something indicates its heading level. e.g. `###` is a subheading under `##`.

### Lists

You can make bullet pointed lists with `*` or `-` -

* one item
* another item

or

- an item
- yet another

Numbered lists will get autonumbered: e.g., look at the source for the below:

1. first
2. second
4. third

## Formatting code examples

"Text" formatting for code etc: `print('hello world')`.

Multiline formatting: three backquotes:
```
print('hello, world')
print(5+5)
```

You can get syntax highlighting by putting 'python' in after the first set of backquotes:
```python
print('hello, world')
print(5+5)
```

## Figures and images

Figures / images in png or jpg formats can be referenced like links, but with a ! in front - `![cat](Black_white_cat_on_fence.jpg)`. Note that the file needs to be referenced by path - it's easiest if it's in the same directory as the notebook!

![cat](Black_white_cat_on_fence.jpg)

## Math

Since Jupyter Notebook is used so much in math and physics and data science, there is special formatting in a system called "LaTeX". Put inline equations inside `$` like so: $ f(x) = 5x$.

Write full equations and equation blocks in double $$:

$$
f(x) = 5(x + 3)
$$
